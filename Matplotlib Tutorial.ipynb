### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#RESIZE GRAPH
plt.figure(figsize=(5,3),dpi=100)
#plt.plot(x,y, label='2x', color='red', linewidth = 1.6, marker = '.', markersize=10, linestyle='--')

#SHORTHAND NOTATION
# fmt = '[color][marker][line]'
plt.plot(x,y,'r.--',label='2x')

#LINE NUMBER 2
x2 = np.arange(0,4.5,0.5)
print(x2)
plt.plot(x2[:6], x2[:6]**2,'b', label='X^2')
plt.plot(x2[5:], x2[5:]**2, 'b--')

plt.title("First plot", fontdict={"fontname": "Comic Sans MS", "fontsize": 20})
plt.xlabel('X Axis', fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

plt.legend()

#SAVE GRAPH
plt.savefig('mygraph.png', dpi=300)

plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)
patterns = ['/', 'O', '-']

for index, bar in enumerate(bars):
    bar.set_hatch(patterns[index])
    
    
#for bar in bars:
#    bar.set_hatch(patterns.pop(0))


#RESIZE
plt.figure(figsize = (6,4))


plt.show()
```

### Line Graph

```python
gas = pd.read_csv('gas_prices.csv')

#RESIZE
plt.figure(figsize=(8,5),dpi=100)

#ADD TITLE
plt.title('Gas Prices over Time (in USD)', fontdict = {
    'fontweight': 'bold',
    'fontsize': 22
})

#plt.plot(gas['Year'], gas['USA'], 'b.-',label='USA')
#plt.plot(gas['Year'], gas['Canada'], 'r.-', label='Canada')
#plt.plot(gas['Year'], gas['South Korea'], 'g.-', label='South Korea')
#plt.plot(gas['Year'], gas['Australia'], 'y.-', label='Australia')

for country in gas:
    if country != 'Year':
        plt.plot(gas['Year'], gas[country], marker = '.', label = country)

plt.xticks(gas['Year'][::3].tolist()+[2011])
plt.ylabel('US Dollars')
plt.xlabel('Year')

plt.legend()


plt.savefig('Gas_price_figure.png', dpi = 300)
plt.show()
```

### Histogram

```python
my_bins = [i for i in range(40,110,10)]
print(my_bins)

plt.hist(fifa['Overall'], bins = my_bins, color = '#abcdef')

plt.xticks(my_bins)
plt.xlabel('Skill Level')
plt.ylabel('Number of Players')
plt.title('Distribution of Player Skills in FIFA 2018')


plt.show()
```

### Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']
colors = ['pink', 'purple']
plt.pie([left, right], labels = labels, colors = colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')
plt.show()
```

### Pie Chart #2

```python
fifa = pd.read_csv('fifa_data.csv')


plt.style.use('ggplot')

fifa['Weight'] = fifa['Weight'].str.replace('lbs' , '')
fifa['Weight'] = pd.to_numeric(fifa['Weight'])
fifa['Weight'] = fifa['Weight'].map(fifa['Weight'] * 0.45359237)
fifa['Weight'] = round(fifa['Weight'],2)

light = fifa.loc[fifa['Weight'] < 65].count()[0]
light_medium = fifa.loc[(fifa['Weight'] >= 65) & (fifa['Weight']<90)].count()[0]
medium = fifa.loc[(fifa['Weight'] >= 90) & (fifa['Weight']<95)].count()[0]
heavy = fifa.loc[(fifa['Weight'] >= 95)].count()[0]


weights = [light, light_medium, medium, heavy]
labels = ['Under 65', '65-90', '90-100', 'Above 100']
plt.pie(weights, labels = labels, autopct='%.2f %%', pctdistance=0.8)
plt.title("FIFA Players' Weight")

plt.show()
```

### Box and Whiskers Chart

```python
plt.style.use('default')

plt.figure(figsize=(5,8),dpi=100)

barcelona = fifa.loc[fifa['Club'] == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa['Club'] == 'Real Madrid']['Overall']
revs = fifa.loc[fifa['Club'] == 'New England Revolution']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'New England Revolution']

boxes = plt.boxplot([barcelona, madrid, revs], labels = labels)

for box in boxes['boxes']:
    box.set(color='pink', linewidth=2)


plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.show()
```
